# Population

- Entire distribution of observations or events
- Costly
- Time consuming
- Good to work with samples

# Sample

- Subset of pupolation's observations
- Represents the characteristics of the population
- The larger the sample size, the closer the statistics to the population
- Observations Should be randomly selected from the populations 

# Central Limit Theorem

- If a large number of observations are selected randomly from some independent population variable, then the distribution of the selected dataset will be normal or nearly normal
- This lets us make inference to the population
- How large is large enough?
    - How accurate you want to be
    - The more normal your current sample dataset, the less extra data you need
    - If your current sample is normal, then 30-40 observation is ok

# Hypothesis Testing

- Hypothesis : A claim that needs testing to prove. It should be well-defined and comparative.
- 2 types:
    - Null Hypothesis: 
        - Represents the stated fact (lets say, x= 5)
        - Represented with `H0`
        - You seek to reject the statement
    - Alternate hypothesis:
        - Everything else not represented by the null hypothesis.
        - Represented with `H1`
- Example:
    - Null hypothesis : Average Ferrary price is same as average car price
    - Alternate hypothesis: Average Ferrary price is not same as average car price
    - We calculate average Ferrary price (`X`)
    - We calculate average car price (`Y`)
    - We judge the difference between X and Y (`Result = X-Y`)
    - If `Result` is too large (significant), we reject the null hypothesis
    - If `Result` is small (not significant), we retain the null hypothesis (FAIL TO REJECT THE NULL HYPOTHESIS)
    - The test how we calculate the difference is called `test statistic`

# T-Test

- Test statistics
- Checks if 2 means are statistically different
- Produces a p value
- p value : tells that if a result is generated by random chance
- p-value cut-offs : 0.01, 0.05
- If p value < cut-off = REJECT THE NULL HYPOTHESIS
- How many tails:
    - Checks only 1 direction (only > or only <) = 1 tailed t-test
    - Checks only 2 directions (both > and <) = 2 tailed t-test
- How many types: 3 types
    - type-1 = same observations, different time    
    - type-2 = different observations, same variance 
    - type-3 = different observations, different variance
- Works when variance is unknown
- Needs 2 range   
- Smaller dataset (< 30) 
- Formula: `=T.TEST(A1:A100,B1:B100, tails, type)`

# Z-Test

- Test statistics
- Checks if 2 means are statistically different
- Produces a z score
- z score (factor) : tells how far away from the mean in the distribution in form of `factor * stdev`
- p-value from z-score : `factor` tells the probability. use chart
- p-value cut-offs : 0.01, 0.05
- If p value < cut-off = REJECT THE NULL HYPOTHESIS
- For 2-tailed test, multiply `Z.TEST()` result by 2
- Works when some statistic is known (eg- mean)
- Bigger dataset (> 30) 
- Needs 1 range      
- measurement is from value to the right in the normal distribution (area).
- Formula: `=Z.TEST(A1:A100, test_stat, stdev)`

# Chi-squared Test

- Comparing samples for meaningful differences (observation before pill consumption, observation after pill consumption)
- provides a p value
- data has to be in 2 independent groups
- dataset must not be too small (at least > 5 is enough) 
- 2 possibilities:
    1. Difference is from random sampling (`H0`)
    2. Difference is due to some reason (`H1`)
- If p value < cut-off(0.05) = REJECT THE NULL HYPOTHESIS

- Formula: `CHITEST(observed_range,expected_range)`
    - `expected_range` : eg- a list full of same average value
